```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as shs


x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y,color="red")
plt.xlabel("Variaveis X")
plt.ylabel("Variaveis Y")
plt.title("Grafico Fds")
plt.show()
```

```python
categorias=["a","b","c","d"]
valores = [10,20,30,40]
plt.bar(categorias, valores)
plt.show()
```

```python
dados = np.random.randn(1000)
plt.hist(dados,bins=20, edgecolor="white")
plt.show()
```

```python
dados = [7,8,9,10,11,12,50]
plt.boxplot(dados)
plt.show()
```
