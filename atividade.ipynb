## Corinthias

```python
df_lista_coringao = pd.read_csv('/Users/jvfra/OneDrive/Documentos/projeto-data-science/todo-poderoso-timao.csv')
df_lista_coringao
```

```python
#11. Qual foi o ano com mais vit√≥rias do Corinthians?

max_vitorias = df_lista_coringao['Vit√≥rias'].idxmax()
numero_vitorias = df_lista_coringao['Vit√≥rias'].max()
ano_com_mais_vitorias = int(df_lista_coringao.loc[max_vitorias, 'Ano'])
ano_com_mais_vitorias
print(f'O ano com mais vit√≥rias do Corinthians foi {ano_com_mais_vitorias}, com {numero_vitorias} vit√≥rias.')
#df_ano_com_mais_vitorias
```

```python
#12. Qual foi o saldo de gols (gols marcados - gols sofridos) em cada ano??

df_lista_coringao['saldo_gols'] = df_lista_coringao['Gols_Marcados'] / df_lista_coringao['Gols_Sofridos']
anos_lista = df_lista_coringao['Ano']
df_lista_coringao[['Ano','saldo_gols']]
```

```python
#13. Qual a m√©dia de posi√ß√£o do Corinthians nos campeonatos?

df_soma_posicao = df_lista_coringao['Posi√ß√£o_Campeonato'].sum() / df_lista_coringao['Posi√ß√£o_Campeonato'].count()
df_posicao_media = df_soma_posicao.mean()
print(f'A m√©dia de posi√ß√£o do Corinthians nos campeonatos √© de {df_posicao_media}.')
```

```python
#14. Qual ano teve o menor n√∫mero de derrotas?

max_derrotas = df_lista_coringao['Derrotas'].idxmax()
numero_derrotas = df_lista_coringao['Derrotas'].max()
ano_com_mais_derrotas = int(df_lista_coringao.loc[max_derrotas, 'Ano'])
ano_com_mais_derrotas
print(f'O ano com mais derrotas do Corinthians foi {ano_com_mais_derrotas}, com {numero_derrotas} derrotas.')
```

```python
#15. Em quais anos o Corinthians marcou mais de 60 gols?

df_lista_conringao_vitorias_60 = df_lista_coringao[df_lista_coringao['Gols_Marcados'] >= 60]
df_lista_conringao_vitorias_60[['Ano','Gols_Marcados']]
```

```python
#16. Qual foi a taxa de vit√≥rias (%) por ano?

df_lista_coringao['taxa_vitorias'] = (df_lista_coringao['Vit√≥rias'] / df_lista_coringao['Jogos']) * 100
df_lista_coringao[['Ano','taxa_vitorias']]
```

```python
#17. Liste os anos em ordem decrescente de pontos (Vit√≥rias = 3 pontos, Empates = 1 ponto).

df_lista_coringao['pontos'] = (df_lista_coringao['Vit√≥rias'] * 3) + df_lista_coringao['Empates'] + (df_lista_coringao['Derrotas'] * -1)
df_lista_coringao_pontos_decrescente = df_lista_coringao[['Ano','pontos']].sort_values(by='pontos', ascending=False)
df_lista_coringao_pontos_decrescente
```

```python
#18. Qual foi a pior posi√ß√£o do Corinthians nesse per√≠odo?

df_index_posicao_menor = df_lista_coringao['Posi√ß√£o_Campeonato'].idxmin()
posicao_menor = df_lista_coringao['Posi√ß√£o_Campeonato'].min()
ano_posicao_menor = int(df_lista_coringao.loc[df_index_posicao_menor, 'Ano'])
print(f'o ano com a menor posi√ß√£o foi o ano de: {ano_posicao_menor}')
```

```python
#19. Qual foi o ano com mais jogos disputados?

df_index_ano_mais_jogos = df_lista_coringao['Jogos'].idxmax()
ano_com_mais_jogos = df_lista_coringao['Jogos'].max()
ano_com_mais_jogos_loc = int(df_lista_coringao.loc[df_index_ano_mais_jogos, 'Ano'])
print(f'o ano com mais jogos foi o ano de: {ano_com_mais_jogos_loc}')


```

```python
#20. Qual a m√©dia de gols sofridos por jogo em cada ano?

df_index_ano_mais_gols_sofridos = df_lista_coringao['Gols_Sofridos'].idxmax()
ano_com_mais_gols_sofridos = df_lista_coringao['Gols_Sofridos'].max()
ano_com_mais_gols_sofridos_loc = int(df_lista_coringao.loc[df_index_ano_mais_gols_sofridos, 'Ano'])
print(f'o ano com mais gols sofridos foi o ano de: {ano_com_mais_gols_sofridos_loc}')
```
